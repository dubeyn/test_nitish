```python
# =========================================================
# SIMPLE RAG (Retrieval-Augmented Generation) EXAMPLE
# =========================================================
#
# PURPOSE:
# This example shows how a basic RAG system works.
#
# WHAT THIS CODE DOES:
# 1. Stores documents
# 2. Converts text into embeddings (vectors)
# 3. Saves embeddings in a vector database
# 4. Searches similar content
# 5. Sends retrieved content to LLM
# 6. Generates final answer
#
# WHY RAG IS USED:
# - Use external knowledge
# - Avoid hallucination
# - Answer from company documents
# - Work with PDFs/manuals/policies
# - Use latest information without retraining
#
# =========================================================


# =========================================================
# IMPORT REQUIRED LIBRARIES
# =========================================================

# SentenceTransformer
# PURPOSE:
# Converts text into embeddings (numerical vectors)
#
# WHY:
# LLM cannot directly search documents efficiently.
# Embeddings help compare semantic meaning.
#
# INSTALL:
# pip install sentence-transformers
#
from sentence_transformers import SentenceTransformer


# NumPy
# PURPOSE:
# Mathematical operations on vectors
#
# WHY:
# Used for similarity calculations
#
# INSTALL:
# pip install numpy
#
import numpy as np


# =========================================================
# STEP 1 : LOAD EMBEDDING MODEL
# =========================================================

# all-MiniLM-L6-v2
#
# PURPOSE:
# Small embedding model from Hugging Face
#
# WHY:
# Converts sentences into vector embeddings
#
# OUTPUT:
# Text → Numerical vector
#
model = SentenceTransformer('all-MiniLM-L6-v2')


# =========================================================
# STEP 2 : CREATE DOCUMENT DATABASE
# =========================================================

# These are sample company documents
#
# In real projects:
# - PDFs
# - Policies
# - Medical records
# - Banking SOPs
# - Research papers
#
documents = [
    "AI governance ensures responsible AI usage.",
    "Machine learning is a subset of artificial intelligence.",
    "RAG combines retrieval and generation.",
    "Fine tuning changes model behavior.",
    "Vector databases store embeddings."
]


# =========================================================
# STEP 3 : CREATE EMBEDDINGS
# =========================================================

# PURPOSE:
# Convert documents into vectors
#
# WHY:
# Similar meanings produce similar vectors
#
# Example:
# "AI governance"
# and
# "Responsible AI"
#
# will have close vector values
#
document_embeddings = model.encode(documents)


# =========================================================
# STEP 4 : USER QUESTION
# =========================================================

# User asks a question
#
query = "What is AI?"


# =========================================================
# STEP 5 : CONVERT QUERY TO EMBEDDING
# =========================================================

# PURPOSE:
# Convert user question into vector
#
query_embedding = model.encode(query)


# =========================================================
# STEP 6 : CALCULATE SIMILARITY
# =========================================================

# PURPOSE:
# Find which document is most similar
#
# METHOD:
# Dot product similarity
#
# WHY:
# Higher similarity = closer meaning
#
similarities = np.dot(document_embeddings, query_embedding)


# =========================================================
# STEP 7 : FIND BEST MATCH
# =========================================================

# np.argmax()
#
# PURPOSE:
# Returns index of highest similarity
#
best_match_index = np.argmax(similarities)


# =========================================================
# STEP 8 : RETRIEVE DOCUMENT
# =========================================================

# PURPOSE:
# Get most relevant document
#
retrieved_document = documents[best_match_index]


# =========================================================
# STEP 9 : BUILD PROMPT FOR LLM
# =========================================================

# PURPOSE:
# Give retrieved context to LLM
#
# WHY:
# This is the core idea of RAG
#
# LLM answers using retrieved knowledge
#
prompt = f"""
Answer the question using the context below.

Context:
{retrieved_document}

Question:
{query}
"""


# =========================================================
# STEP 10 : FINAL OUTPUT
# =========================================================

print("USER QUESTION:")
print(query)

print("\nRETRIEVED DOCUMENT:")
print(retrieved_document)

print("\nPROMPT SENT TO LLM:")
print(prompt)


# =========================================================
# EXPECTED OUTPUT
# =========================================================
#
# USER QUESTION:
# What is RAG?
#
# RETRIEVED DOCUMENT:
# RAG combines retrieval and generation.
#
# =========================================================
# REAL ENTERPRISE RAG ARCHITECTURE
# =========================================================
#
# User Question
#       ↓
# Embedding Model
#       ↓
# Vector Database
#       ↓
# Similar Documents
#       ↓
# LLM Prompt
#       ↓
# Final AI Response
#
# =========================================================
# POPULAR VECTOR DATABASES
# =========================================================
#
# 1. Pinecone
# PURPOSE:
# Cloud vector database
#
# 2. ChromaDB
# PURPOSE:
# Lightweight local vector database
#
# 3. FAISS
# PURPOSE:
# Fast similarity search by Meta
#
# 4. Weaviate
# PURPOSE:
# Enterprise semantic search
#
# =========================================================
# WHY COMPANIES USE RAG
# =========================================================
#
# - Chat with PDFs
# - Banking assistants
# - AI governance systems
# - Medical knowledge search
# - Legal assistants
# - Internal company chatbot
# - Research systems
#
# =========================================================
# LIMITATION OF THIS SIMPLE EXAMPLE
# =========================================================
#
# This is educational only.
#
# Real projects additionally use:
#
# - Chunking
# - Vector databases
# - Metadata filtering
# - LangChain
# - LlamaIndex
# - OpenAI/HuggingFace LLMs
# - Hybrid search
# - Re-ranking
#
# =========================================================
```
